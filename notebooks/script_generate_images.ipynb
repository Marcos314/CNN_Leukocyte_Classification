### Criando matriz de posição (PM)

```python
#seq = "AACCGGTT"
seq = array_seq[0]
def split(seq): 
    return [char for char in seq]
def getA(listSeq):
    posA = []
    for nucl in listSeq:
        if nucl == "A":
            posA.append(1)
        else:
            posA.append(0)
    return posA

def getC(listSeq):
    posC = []
    for nucl in listSeq:
        if nucl == "C":
            posC.append(1)
        else:
            posC.append(0)
    return posC

def getG(listSeq):
    posG = []
    for nucl in listSeq:
        if nucl == "G":
            posG
def getT(listSeq):
    posT = []
    for nucl in listSeq:
        if nucl == "T":
            posT.append(1)
        else:
            posT.append(0)
    return posT    

import pandas as pd
matrix = pd.DataFrame([getA(split(seq)), getC(split(seq)), getG(split(seq)), getT(split(seq))])
print(matrix)

import seqlogo
ppm = seqlogo.Ppm(matrix)
ppm
```
