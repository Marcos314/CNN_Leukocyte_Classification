## Transformação do dataset tarbase_data.csv
- Selecionar apenas as espécies de interesse
- Adicionar uma coluna com as sequencias de cada mirna

```
# Import
import pandas as pd
import numpy as np
```

```
df = pd.read_csv('/content/drive/My Drive/TCC - 2020/tarbase_data.csv', sep='\t')
```

```
df.shape
```

```
df.head()
```

```
# Filtrando as 5 espécies observadas
df = df.loc[df['species'].isin(['Arabidopsis thaliana','Oryza sativa','Zea mays','Medicago truncatula','Vitis vinifera'])]
```

```
df.shape
```

```
df.head()
```

```
# Eliminando algumas colunas
df = df.drop(columns=['category','method','positive_negative','direct_indirect','up_down','condition','cell_line','tissue'])
df.head()
```

```
df.shape
```

```
# Verificando quantos miRNA  únicos existem (desconsidera valores repetidos)
df['mirna'].nunique()
```

```
# Verificando a frequência de cada miRNA
df['mirna'].value_counts()
```

## Arquivos fasta (.fa)



```
!pip install biopython
```

```
from Bio import SeqIO

lista_seq = []
lista_id = []

for fasta in SeqIO.parse('/content/drive/My Drive/TCC - 2020/mature.fa','fasta'):
    
    lista_id.append(fasta.id)
    lista_seq.append(str(fasta.seq))
    
    

    #print(fasta.id)
    #print(fasta.seq)

```

```
len(lista_id)
```

```
len(lista_seq)
```

```
print(lista_seq)
```

```
# Criando um dicionário a partir das listas
my_dict = {'id':lista_id,'sequencia':lista_seq}
```

```
print(my_dict)
```

```
# Criar um dataframe a partir do dicionário
df_seq = pd.DataFrame(data=my_dict)
```

```
df_seq.head()
```

## Fazendo um merge (inner join) entre o dataframe com os genesId e o dataframe com as sequências

```
new_df = pd.merge(df, df_seq,left_on='mirna', right_on='id', how='left')
```

```
new_df.head()
```

```
# Verificando valores ausentes no dataset
new_df.isnull().sum()
```

```
new_df['sequencia'].describe()
```

```
new_df['mirna'].describe()
```

```
new_df['id'].describe()
```

**Podemos observar a partir dos valores unicos descritos acima que existem 316 mirna's unicos, porém só 177 sequências unicas. Logo, 139 sequências estão associadas a mais de um mirna.**
